```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lid_openness = pd.read_csv("../data/20240827_lid_openness_alphafold.csv")
```

```python
e_coli_closed = 8.8 # 1ake
e_coli_open = 24.4 # 4ake
```

```python
# vertical histogram of lid openness
fig = plt.figure(figsize=(2, 6))
plt.hist(lid_openness["lid_openness"], bins=50, color="blue", orientation="horizontal", edgecolor="black")
plt.axhline(e_coli_closed, color="red", linestyle="--")
plt.axhline(e_coli_open, color="red", linestyle="--")
plt.xlabel("Frequency")
plt.ylabel("Lid openness")
plt.rcParams['pdf.fonttype'] = 42
plt.savefig("../data/fig_pdfs/lid_openness_vertical_hist.pdf")
```
